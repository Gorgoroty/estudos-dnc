###Monte um código onde a primeira pergunta será "Qual curso você quer escolher (1 =  Formação_em_dados, 2 = Formação_em_projetos, 3 = Formação_em_marketing, 4=  Formação_em_produto, 5 = Formação_em_tecnologia)?"E de acordo com a escolha, ele entre em um if e emita a seguinte mensagem padrão:

"Você escolheu o curso Formação em XXXXX no valor de R$ valor que começará no  próximo mês".Caso o usuário digite um valor diferente dos dispostos, emita a seguinte mensagem  "Digite um valor correspondente a um curso de nossas formações \n1-Formação em  dados \n2-Formação em projetos\n3-Formação em marketing\n4-Formação em  produto\n5-Formação em tecnologia"




```python
cursos={'Formação em Dados': 20000.00,
        'Formação em Projetos': 15000.00,
        'Formação em Marketing': 13000.00,
        'Formação em Produto': 18000.00,
        'Formação em Tecnologia': 19000.00}


opc=int(input('Qual curso você quer escolher (1 = Formação_em_dados, 2 = Formação_em_projetos, 3 = Formação_em_marketing, 4= Formação_em_produto, 5 = Formação_em_tecnologia)?\n'))
texto_err='Digite um valor correspondente a um curso de nossas formações \n1-Formação em dados \n2-Formação em projetos\n3-Formação em marketing\n4-Formação em produto\n5-Formação em tecnologia \n'
opções=(1,2,3,4,5)
if opc not in opções:
  while opc not in opções:
    opc=int(input(texto_err))
    if opc in opções:
      break


if opc==1:
  print('Você escolheu o curso em Formação em Dados no valor de R${}'.format(cursos['Formação em Dados']))
elif opc==2:
  print('Você escolheu o curso em Formação em Projetos no valor de R${}'.format(cursos['Formação em Projetos']))
elif opc==3:
  print('Você escolheu o curso em Formação em Marketing no valor de R${}'.format(cursos['Formação em Marketing']))
elif opc==4:
  print('Você escolheu o curso em Formação em Produto no valor de R${}'.format(cursos['Formação em Produto']))
elif opc==5:
  print('Você escolheu o curso em Formação em Tecnologia no valor de R${}'.format(cursos['Formação em Tecnologia']))

```

##**Questão 5**. A DNC pretende faturar 2 milhões em 1 mês. Faça a leitura da quantidade de vendas de cada formação e calcule a receita total de vendas, a receita media geral e verifique se eles bateram a meta mensal utilizando as seguintes mensagens 
###Se bateram: "Parabéns!!! A DNC bateu a meta do mês"
###Se não bateram: "Esse mês não batemos a meta, porém mês que vem vamos vir mais forte pra recuperar!!!"


(Utilize as informações de preço do exercício acima)

```python
vendas=list()
i=0
soma=0
for keys in cursos:
  vendas.append(int(input('Quantos você vendeu de {} ?'.format(keys))))
  soma+=vendas[i]*cursos[keys]
  i+=1
if soma>=2000000.00:
  print('Parabéns!!! A DNC bateu a meta do mês! Vocês venderam R${:.2f}'.format(soma))
else:
  print('Esse mês não batemos a meta. Infelizmente só vendemos R${:.2f}. Porém mês que vem vamos vir mais forte pra recuperar!!!'.format(soma))
```

##**Questão 6**. O aluno da DNC precisa fazer um código que faça a somatória de uma  sequência de números um por um, informando no começo do código qual a posição a somatória  deve parar. Monte uma estrutura de repetição que resolva esse problema.

```python
n=int(input('Até que número deseja somar ? \n'))
soma=0
for i in range(n,0,-1):
  soma+=i
print('O valor dos números somados até {} é de {}'.format(n,soma))
```
